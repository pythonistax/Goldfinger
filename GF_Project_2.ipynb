```python
import os 
import pandas as pd

current_dir = os.getcwd()
files = os.listdir(current_dir)
for file in files:
    if "GF_Affiliate_PnL_Export_" in file and file.endswith('.csv'):
        print(file)
        file_name = file
        
df_pnl = pd.read_csv(file_name)
df_pnl = df_pnl[df_pnl['Customers'] > 25].sort_values(by='Customers', ascending=False)
df_pnl.head()








```
